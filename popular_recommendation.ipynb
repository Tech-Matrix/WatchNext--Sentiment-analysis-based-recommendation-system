```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\Swarupa V\Desktop\watchnext\dataset\datasets_1474_2639_IMDB-Movie-Data.csv')
```

```python
df.head()
```

```python
df['Metascore'].fillna(1,inplace=True)
df.head(30)
```

```python
action=[]
drama=[]
thriller=[]
adventure=[]
comedy=[]
horror=[]
biography=[]
animation=[]
scifi=[]
romance=[]
others=[]
```

```python
#to get a list of all movies in all genre
genres=list(df["Genre"].values)
movies=list(df["Title"])
i=0
for genre,movie in zip(genres,movies):
    lst=genre.split(',')
    for g in lst:
        if(g=="Action" or g=='War'):
            action.append(movie)
        if(g=='Drama' or g=='Musical' or g=='Family'):
            drama.append(movie)
        if(g=='Thriller' or g=='Crime' or g=='Mystery'):
            thriller.append(movie)
        if(g=='Adventure'):
            adventure.append(movie)
        if(g=='Comedy'):
            comedy.append(movie)
        if(g=='Horror'):
            horror.append(movie)
        if(g=='Biography' or g=='History'):
            biography.append(movie)
        if(g=='Animation' or g=='Fantasy'):
            animation.append(movie)
        if(g=='Sci-Fi'):
            scifi.append(movie)
        if(g=='Romance'):
            romance.append(movie)
        else:
            others.append(movie)
        
```

```python
print(scifi)
```

```python
#surprise for scifi
rating=[]
for movie in scifi:
    r=int(df.loc[df['Title']==movie,'Rating'])
    m=int(df.loc[df['Title']==movie,'Metascore'])
    score=(r*0.3)+(m*0.7)
    v=int(df.loc[df['Title']==movie,'Votes'])
    if(r>7):
        movie_score=(v/20000)*score #opularity score where movies with rating more than 7  and more than 200000 votes lead to a better score
        rating.append(movie_score)
    else:
        movie_score=score+0.07*v
        rating.append(movie_score)
```

```python
scifi_final=list(zip(rating,scifi))
scifi_final[45][1] 
```
