```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\Swarupa V\Desktop\watchnext\dataset\datasets_1474_2639_IMDB-Movie-Data.csv')
```

```python
df.head()
```

```python
df['Metascore'].fillna(1,inplace=True)
df.head(3)
```

```python
action=[]
drama=[]
thriller=[]
adventure=[]
comedy=[]
horror=[]
biography=[]
animation=[]
scifi=[]
romance=[]
others=[]
```

```python
genres=list(df["Genre"].values)
movies=list(df["Title"])
i=0
for genre,movie in zip(genres,movies):
    lst=genre.split(',')
    for g in lst:
        if(g=="Action" or g=='War'):
            action.append(movie)
        if(g=='Drama' or g=='Musical' or g=='Family'):
            drama.append(movie)
        if(g=='Thriller' or g=='Crime' or g=='Mystery'):
            thriller.append(movie)
        if(g=='Adventure'):
            adventure.append(movie)
        if(g=='Comedy'):
            comedy.append(movie)
        if(g=='Horror'):
            horror.append(movie)
        if(g=='Biography' or g=='History'):
            biography.append(movie)
        if(g=='Animation' or g=='Fantasy'):
            animation.append(movie)
        if(g=='Sci-Fi'):
            scifi.append(movie)
        if(g=='Romance'):
            romance.append(movie)
        else:
            others.append(movie)
        
```

```python
#popular for scifi
rating1=[]
for movie in scifi:
    r=int(df.loc[df['Title']==movie,'Rating'])
    m=int(df.loc[df['Title']==movie,'Metascore'])
    v=int(df.loc[df['Title']==movie,'Votes'])
    score=(r)+(m/10)
    if(r>7):
        movie_score=((20000-v)/10000)*score# for movies that have rating >7 and less number of votes, greater score is given
        rating1.append(movie_score)
    else:
        movie_score=score-(v/10000)
        rating1.append(movie_score)
```

```python
#popular for fantasy,animation
rating6=[]
for movie in animation:
    r=int(df.loc[df['Title']==movie,'Rating'])
    m=int(df.loc[df['Title']==movie,'Metascore'])
    v=int(df.loc[df['Title']==movie,'Votes'])
    score=(r)+(m/10)
    if(r>7):
        movie_score=((20000-v)/10000)*score
        rating6.append(movie_score)
    else:
        movie_score=score-(v/10000)
        rating6.append(movie_score)
```

```python
scifi_final=list(zip(rating1,scifi))
ani_final=list(zip(rating6,animation)) 
```

```python
scifi_final.sort()
ani_final.sort()
```
